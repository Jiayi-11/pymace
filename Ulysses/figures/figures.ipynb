```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data points
baseline1 = [1.92, 1.3, 1.34, 1.73, 1.62, 2.1, 1.99, 0.84, 1.12, 1.11]
baseline2 = [2.29, 1.39, 2.96, 2.46, 1.73, 4.07, 2.18, 2.03, 2.12, 1.8]
our_model = [0.48, 0.49, 0.85, 1.08, 0.58, 0.63, 0.57, 0.45, 0.52, 0.44]

# Combine data into a DataFrame
data = {
    "scenario": ["our_model"] * 10 + ["baseline1"] * 10 + ["baseline2"] * 10,
    "value": our_model + baseline1 + baseline2
}

df = pd.DataFrame(data)

# Define color palette
color_map = {
    "our_model": "C0",
    "baseline1": "C1",
    "baseline2": "C2"
}

# Create the plot
fig, ax = plt.subplots(figsize=(6, 5))

# Boxplot
sns.boxplot(
    x="scenario",
    y="value",
    data=df,
    palette=color_map,
    width=0.4,  # Reduce the width to make the boxes narrower
    ax=ax
)

ax.set_xlabel("Scenario")
ax.set_ylabel("Completion time (hour)")

# Setting y-axis limit
plt.ylim(0, 3.5)

# Adjust layout
plt.tight_layout()
plt.grid(True)
plt.savefig("new_boxplot.svg", format="svg", bbox_inches="tight")

# Show the plot
plt.show()
```
