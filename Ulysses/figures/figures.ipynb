```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
baseline1 = [1.92, 1.3, 1.34, 1.73, 1.62, 2.1, 1.99, 0.84, 1.12, 1.11]
baseline2 = [2.23, 2.23, 1.94, 2.21, 2.21, 2.96, 2.12, 1.68, 1.94, 1.65]
our_model = [0.53, 1.04, 0.49, 0.36, 0.42, 0.47, 0.41, 0.57, 0.53, 0.55]

# Calculating means and standard deviations for error bars
mean1 = np.mean(baseline1)
std1 = np.std(baseline1)

mean2 = np.mean(baseline2)
std2 = np.std(baseline2)

mean3 = np.mean(our_model)
std3 = np.std(our_model)

# Indices for plotting
indices = np.arange(1, len(baseline1) + 1)

# Plotting with lines connecting the points
plt.figure(figsize=(10, 5))
plt.errorbar(indices, baseline1, yerr=std1,
             fmt='o-', label='Baseline 1', capsize=5)
plt.errorbar(indices, baseline2, yerr=std2,
             fmt='^-', label='Baseline 2', capsize=5)
plt.errorbar(indices, our_model, yerr=std3,
             fmt='*-', label='our_model', capsize=5)

# Setting x-axis ticks to be in units of 1
plt.xticks(indices)

plt.title('Error Bar Plot of time to complete the inspection mission')
plt.xlabel('Number of runs')
plt.ylabel('Completion time (hour)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
baseline1 = [1.92, 1.3, 1.34, 1.73, 1.62, 2.1, 1.99, 0.84, 1.12, 1.11]
baseline2 = [2.23, 2.23, 1.94, 2.21, 2.21, 2.96, 2.12, 1.68, 1.94, 1.65]
our_model = [0.53, 1.04, 0.49, 0.36, 0.42, 0.47, 0.41, 0.57, 0.53, 0.55]

# All data points
data = [our_model, baseline1, baseline2]

# Calculating means for each dataset
means = [np.mean(our_model), np.mean(baseline1), np.mean(baseline2)]
print(means)

# Plotting the boxplot
plt.figure(figsize=(7, 5))
box = plt.boxplot(data, patch_artist=True, notch=False, vert=True, showmeans=True, meanline=True)

# Adding means to the boxplot
#colors = ['blue', 'orange', 'green']
#for i in range(len(means)):
#    plt.plot(i + 1, means[i], 's', color=colors[i], label=f'Mean of {"Baseline 1" if i == 0 else "Baseline 2" if i == 1 else "Our Model"}')
#    plt.text(i + 1.1, means[i], f'{means[i]:.3f}', color=colors[i], va='center')

# Setting colors for each box
colors = ['#2ca02c', '#1f77b4', '#ff7f0e']  # green, blue, orange


for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor(color)


for i, color in enumerate(colors):
    plt.setp(box['boxes'][i], color=color)
    plt.setp(box['whiskers'][2*i:2*(i+1)], color=color)
    plt.setp(box['caps'][2*i:2*(i+1)], color=color)
    plt.setp(box['medians'][i], color=color)
    plt.setp(box['means'][i], markerfacecolor=color, markeredgecolor=color)
    plt.setp(box['fliers'][i], markerfacecolor=color, markeredgecolor=color)

# Setting x-axis labels
plt.xticks([1, 2, 3], ['Our Model', 'Baseline 1', 'Baseline 2'])

# Setting y-axis limit
plt.ylim(0, 3.5)

plt.title('Boxplot of time to complete the inspection mission')
plt.xlabel('Scenarios')
plt.ylabel('Completion time (hour)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data points
baseline1 = [1.92, 1.3, 1.34, 1.73, 1.62, 2.1, 1.99, 0.84, 1.12, 1.11]
baseline2 = [2.23, 2.23, 1.94, 2.21, 2.21, 2.96, 2.12, 1.68, 1.94, 1.65]
our_model = [0.86, 0.9, 0.52, 0.56, 0.51, 0.59, 0.52, 0.39, 0.47, 0.48]

# All data points
data = [our_model, baseline1, baseline2]

# Creating a DataFrame for seaborn
import pandas as pd

df = pd.DataFrame({
    'Completion time (hour)': our_model + baseline1 + baseline2,
    'Scenario': ['Our Model'] * len(our_model) + ['Baseline 1'] * len(baseline1) + ['Baseline 2'] * len(baseline2)
})

# Plotting the violin plot
plt.figure(figsize=(7, 5))
sns.violinplot(x='Scenario', y='Completion time (hour)', data=df, inner=None, palette=['green', 'blue', 'orange'])

# Calculating means for each dataset
means = [np.mean(our_model), np.mean(baseline1), np.mean(baseline2)]

# Adding means to the violin plot
for i, mean in enumerate(means):
    plt.plot(i, mean, 's', color='red')
    plt.text(i, mean, f'{mean:.3f}', color='red', va='center', ha='center')

# Setting y-axis limit
plt.ylim(0, 3.5)

plt.title('Violin Plot of time to complete the inspection mission')
plt.xlabel('Scenarios')
plt.ylabel('Completion time (hour)')
plt.grid(True)
plt.show()
```
